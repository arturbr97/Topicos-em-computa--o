6) Aplique pelo menos 3 m√©todos diferentes para binarizar uma imagem ('skimage.filters').

```python
isodata = threshold_isodata(gray_imagem)
otsu = threshold_otsu(gray_imagem)
li = threshold_li(gray_imagem)
yen = threshold_yen(gray_imagem)
black_white_isodata = gray_imagem < isodata
black_white_otsu = gray_imagem < otsu 
black_white_li = gray_imagem < li
black_white_yen = gray_imagem < yen
```
